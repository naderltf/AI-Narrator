```python
import torch
from parler_tts import ParlerTTSForConditionalGeneration
from transformers import AutoTokenizer
import soundfile as sf
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"

model = ParlerTTSForConditionalGeneration.from_pretrained("parler-tts/parler-tts-mini-v1").to(device)
tokenizer = AutoTokenizer.from_pretrained("parler-tts/parler-tts-mini-v1")
```

```python
import PyPDF2

def extract_text_from_pdf(pdf_path, start_page, end_page):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)

        # Ensure end_page does not exceed the total number of pages
        end_page = min(end_page, num_pages)

        for page_num in range(start_page - 1, end_page):  # Pages are 0-indexed
            text += reader.pages[page_num].extract_text() + "\n"

    return text

# Usage
pdf_path = "/content/short-stories-for-children-ingles-primaria-continuemos-estudiando.pdf"  # Change this to your PDF file path
extracted_text = extract_text_from_pdf(pdf_path, start_page=5, end_page=12)
print(extracted_text)
```

```python
storybook = """
Once upon a time, in a quiet little village, a curious girl named Lila discovered a tiny golden door at the base of an ancient oak tree. It was no bigger than her hand, yet when she turned the miniature knob, the door grew large enough for her to step through.

On the other side, Lila found herself in a land of shimmering rivers that flowed backward, trees that whispered riddles, and floating islands covered in candy-colored grass. A fluffy white fox wearing a pocket watch scurried past her, muttering, "Oh dear, oh dear, I'm almost late for the Feather Festival!"

Intrigued, Lila followed the fox to a grand castle made of woven vines and glowing mushrooms. Inside, a council of owls in tiny spectacles debated over the best way to make the sky sparkle at night. "Stars or fireflies?" they asked Lila.

She thought for a moment and smiled. "Why not both?" The owls hooted in delight and flapped their wings, sending a shower of twinkling fireflies into the sky, where they nestled among the stars.

Just as the Feather Festival began, with birds of every color performing dazzling aerial dances, Lila felt a gentle tug. The golden door had reappeared behind her. She knew it was time to go home.

With one last look at the wondrous land, she stepped through the door and found herself back in her village, the ancient oak tree standing silent as ever. But in her pocket, she found a single glowing feather—a reminder that magic was always just a step away.
"""
```

```python
import re

# Clean the text (remove extra spaces, newlines, etc.)
cleaned_text = re.sub(r'\s+', ' ', storybook).strip()

# Split the text into chunks (e.g., 200 words per chunk)
def split_text(text, max_words=200):
    words = text.split()
    chunks = []
    current_chunk = ""
    for word in words:
        if len(current_chunk.split()) < max_words:
            current_chunk += " " + word
        else:
            chunks.append(current_chunk.strip())
            current_chunk = word
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks

text_chunks = split_text(cleaned_text)
print(f"Split text into {len(text_chunks)} chunks.")
```

```python
description = "Jenna's voice is clear and warm, with a gentle, expressive tone. She speaks slowly and articulately, maintaining a calm and soothing rhythm in a quiet environment, making it perfect for storytelling and creating an immersive listening experience."
input_ids = tokenizer(description, return_tensors="pt").input_ids.to(device)
audio_files = []
for i, chunk in enumerate(text_chunks):
    print(f"Processing chunk {i+1}/{len(text_chunks)}...")
    prompt_input_ids = tokenizer(chunk, return_tensors="pt").input_ids.to(device)
    generation = model.generate(input_ids=input_ids, prompt_input_ids=prompt_input_ids)
    audio_arr = generation.cpu().numpy().squeeze()
    audio_file = f"chunk_{i+1}.wav"
    sf.write(audio_file, audio_arr, model.config.sampling_rate)
    audio_files.append(audio_file)
    print(f"Saved {audio_file}")
```

```python
import torch
from parler_tts import ParlerTTSForConditionalGeneration
from transformers import AutoTokenizer
import soundfile as sf

# Set up device
device = "cuda:0" if torch.cuda.is_available() else "cpu"

# Load model and tokenizer
model = ParlerTTSForConditionalGeneration.from_pretrained("parler-tts/parler-tts-mini-multilingual-v1.1").to(device)
tokenizer = AutoTokenizer.from_pretrained("parler-tts/parler-tts-mini-multilingual-v1.1")
description_tokenizer = AutoTokenizer.from_pretrained(model.config.text_encoder._name_or_path)

# Define description for the voice
description = "Christine's voice is clear and warm, with a gentle, expressive tone. She speaks slowly and articulately, maintaining a calm and soothing rhythm in a quiet environment, making it perfect for storytelling and creating an immersive listening experience."

# Tokenize the description
input_ids = description_tokenizer(description, return_tensors="pt").input_ids.to(device)

# Generate audio for each chunk
audio_files = []
for i, chunk in enumerate(text_chunks):
    print(f"Processing chunk {i+1}/{len(text_chunks)}...")
    prompt_input_ids = tokenizer(chunk, return_tensors="pt").input_ids.to(device)
    generation = model.generate(input_ids=input_ids, prompt_input_ids=prompt_input_ids)
    audio_arr = generation.cpu().numpy().squeeze()
    audio_file = f"chunk_{i+1}.wav"
    sf.write(audio_file, audio_arr, model.config.sampling_rate)
    audio_files.append(audio_file)
    print(f"Saved {audio_file}")
```
